```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vcfcalls = pd.read_csv("results/all_samples_with_var_comparison", header=None, sep="\t", 
                       names=["Padding", "Sample", "Num of var", "Var missing compared to 300-600"])
```

```python
vcfcalls
```

```python
plt.figure(figsize=(12,7))
varplot = sns.stripplot(x="Sample", y="Num of var", hue="Padding",
                   data=vcfcalls, palette="pastel", size=10, dodge=True)
varplot.set_xticklabels(varplot.get_xticklabels(),rotation=25)
varplot.axes.set_title("Variant numbers lancet based on window and padding size\n",fontsize=20)
plt.savefig('results/var_comparison.png')
```
